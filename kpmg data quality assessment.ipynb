## Join Analysis

```python
#Joining Transactions and Customers
df.sort_values("customer_id")
```

```python
df_demo
```

```python
trans_demo=pd.merge(df,df_demo,on=["customer_id"],indicator=True,how="outer")
```

```python
trans_demo
```

```python
trans_demo["_merge"].value_counts()
```

```python
trans_demo[trans_demo["_merge"]=="left_only"] #no such customer_id exists in demographic table
```

```python
trans_demo[trans_demo["_merge"]=="right_only"] #none of these customers made any transactions
```

```python
#joining Customers demograohics and customer addresses
demo_adrs=pd.merge(df_demo,cust_ads,on=["customer_id"],indicator=True,how="outer")
```

```python
demo_adrs
```

```python
demo_adrs[demo_adrs["_merge"]=="right_only"] #no demographics for such customers
```

```python
demo_adrs[demo_adrs["_merge"]=="left_only"] #no address details exist for such customers
```

```python
trans_adrs=pd.merge(df,cust_ads,on=["customer_id"],indicator=True,how="outer")
```

```python
trans_adrs[trans_adrs["_merge"]=="left_only"].count() #missing from addresses
```

```python
trans_adrs[trans_adrs["_merge"]=="right_only"] #missing from transactions
```
